```python
from sklearn.model_selection import train_test_split
```

```python
drp = telecom[['state','area code','phone number','international plan','voice mail plan','churn']]
X= telecom.drop(drp,1)
y= telecom.churn
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
import numpy as np # linear algebra
import matplotlib.pyplot as plt

error_rate = []

for i in range(1,40):

    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))

plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='solid', marker='*',
         markerfacecolor='red', markersize=10)
plt.title('Elbow method to determine K value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
clf = KNeighborsClassifier(6)
clf.fit(X_train, y_train)
```

```python
predicted = clf.predict(X_test)
predicted
```

```python
predicted.shape
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,predicted)
print cm
```

```python
from sklearn import metrics
accuracy = metrics.accuracy_score(predicted,y_test)
Classification_error_rate=1-accuracy
print("KNN Accuracy for the data is : %s" % "{0:.1%}".format(accuracy))
print("KNN error rate for the data is : %s" % "{0:.1%}".format(Classification_error_rate))
```

```python
from sklearn.metrics import classification_report
print classification_report(y_test, predicted)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
y_pred = logreg.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(X_test, y_test)))
```

```python
from sklearn import model_selection
from sklearn.model_selection import cross_val_score
kfold = model_selection.KFold(n_splits=10, random_state=7)
modelCV = LogisticRegression()
scoring = 'accuracy'
results = model_selection.cross_val_score(modelCV, X_train, y_train, cv=kfold, scoring=scoring)
print("10-fold cross validation average accuracy: %.3f" % (results.mean()))
```
